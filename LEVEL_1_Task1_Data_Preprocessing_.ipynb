# Task 1: Data Preprocessing for Machine Learning
I'm starting with data preprocessing because honestly, if your data is messy,
no model in the world will save you.

Dataset: iris.csv
Tools: Python, pandas, scikit-learn

```python
#IMPORTING LIBRARIES
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
# -----------------------------
# 1. Load the dataset
# -----------------------------
df = pd.read_csv("iris.csv")
```

```python
# Quick look at the data
print("First 5 rows of the dataset:")
df.head()
```

```python
print("Last 5 rows of the dataset:")
df.tail()
```

```python
# -----------------------------
# 2. Handle Missing Data
# -----------------------------
# Note: Iris dataset usually has no missing values,
# but we demonstrate the process for completeness.

# Check for missing values
print("\nMissing values per column:")
print(df.isnull().sum())
```

If missing values existed, we could fill them:
df['sepal_length'].fillna(df['sepal_length'].mean(), inplace=True)
Or drop rows with missing values:
df.dropna(inplace=True)

```python
# -----------------------------
# 3. Encode Categorical Variables
# -----------------------------
# 'species' is categorical, so we encode it.
label_encoder = LabelEncoder()
df['species'] = label_encoder.fit_transform(df['species'])

print("\nEncoded species values:")
print(df['species'].unique())  # 0,1,2 for setosa, versicolor, virginica
```

```python
# -----------------------------
# 4. Separate Features and Target
# -----------------------------
X = df.drop('species', axis=1)  # Features
y = df['species']               # Target
```

```python
# -----------------------------
# 5. Normalize / Standardize Features
# -----------------------------
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# -----------------------------
# 6. Split into Training and Testing Sets
# -----------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)
```

```python
print("\nTraining set size:", X_train.shape)
print("Testing set size:", X_test.shape)
```
